```python
# if i get 200 students today, i will teach 
# 120 min of last day else i will teach 60 min only 
```

```python
live = 200 

if live == 200 :
  print("I will teach you guys 120 Min on day 5 ")
else:
  print("I will teach you guys 60 Min on day 5 ")
```

```python
# "==" stands for "Equal to ?" 
# ">=" stand for "Greater than equal to ?" 
# "<=" stand for "less than equal to ?"
# "!=" stand for "Not Equal to ?"
# "<" stand for "Less than ?"
# ">" stand for "Greater than ?"
```

```python
live = 250 

if live != 200 :
  print("I will teach you guys 120 Min on day 5 ")
else:
  print("I will teach you guys 60 Min on day 5 ")
```

```python
# If i get 200 live count i will teach 120 min 
# else if i get 150 live count i will teach 90 min 
# else i will teach 60 min only 
```
